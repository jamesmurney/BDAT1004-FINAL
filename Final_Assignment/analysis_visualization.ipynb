```python
# import essential libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = r'C:\Users\Abhishek\Desktop\AmphibiansData.csv'      # read csv file
data = pd.read_csv(url)
species = pd.DataFrame(data)                               # convert csv file to dataframe
species.head()                                             # show top 5 data rows

```

```python
# count no. of species belonging to differnt IUCN Red list status

red_list = species['IUCN Red List Status'].value_counts()
red_list
```

```python
categories = ['Least Concern(LC)', 'Vulnerable(VU)','Near Threatened(NT)','Endangered(EN)',
              'VU Provisional(VU)','Extinct-Wild(EW)','LC Provisional(LC)','Extinct(EX)']      
num_species = [104, 24, 17, 12, 2, 1, 1, 1]
colors = ['#FFAE39','#2AC280','#DC5B3B','#52E397','#C5CBA3','#9CD5F6','#6E50D9','#9A5E59','#9BC8F5','#0BBC66',
           '#FF3D52','#3B19CE','#25974F','#1EAD79','#375E4C','#43C711','#A4CCD9','#AFF373','#392D6A','#792D86','#9CA54B',
           '#7F03BE','#D5EC95','#FCE2A5']
```

```python
# Create barchart showing number of species in different IUCN list

sns.set(style="darkgrid")
fig = plt.subplots(figsize=(16, 10))
barplot = plt.bar(categories, num_species, color = colors)
for bar in barplot:
    bars = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, bars, int(bars), va='bottom') #va: vertical alignment y positional argument
    
plt.title("Species Status Under IUCN Red List", size = 20)
plt.xlabel('IUCN Red List Status', size = 15, labelpad = 14)
plt.ylabel('Number Of Species', size = 15, labelpad = 10)


plt.savefig(r'C:\Users\Abhishek\Desktop\RedList_BarGraph.png')
```

```python
labels = categories
red_list.value_counts().plot(kind='pie',labels=labels,figsize = (14,10),
                             explode = (0.1, 0.1, 0, 0, 0, 0),
                           autopct = '%1.1f%%', shadow=True,
                           radius = 1, startangle=90,
                           textprops = {"fontsize":15})

plt.title('Species Status Under IUCN Red List',size = 20)
plt.xlabel('Status', size = 18, labelpad = 12)

plt.savefig(r'C:\Users\Abhishek\Desktop\piechart1.png')
```
